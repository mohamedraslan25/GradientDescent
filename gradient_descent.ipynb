```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    theta_0 = 0
    theta_1 = 0
    m = len(x)
    alpha = 0.01
    iterations = 1000
    
    for i in range(iterations):
        y_predict = theta_0 + x * theta_1
        cost = (1/m) * sum([val ** 2 for val in (y - y_predict)])
        theta_0_der = -(2/m) + sum(y - y_predict)
        theta_1_der = -(2/m) + sum((y - y_predict) * x)
        theta_0 = theta_0 - alpha * theta_0_der
        theta_1 = theta_1 - alpha * theta_1_der
        print('iteration: {0}, coefficient: {1}, intercept: {2}, cost: {3}'.format(i, theta_1, theta_0, cost))
```

```python
def gradient_descent(X, y, learning_rate=0.01, n_iterations=1000):
    m = len(X)  
    X_b = np.c_[np.ones((m, 1)), X]
    
    theta = np.random.randn(X_b.shape[1], 1)
    
    for iteration in range(n_iterations):
        predictions = X_b.dot(theta)
        
        gradients = (2 / m) * X_b.T.dot(predictions - y)
        
        theta = theta - learning_rate * gradients
    
    return theta

def plot_regression(X, y, theta):
    plt.scatter(X, y, color='blue', label='Data points')
    
    X_b = np.c_[np.ones((len(X), 1)), X]
    predicted_values = X_b.dot(theta)
    
    plt.plot(X, predicted_values, color='red', label='Predicted line')
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title("Linear Regression Fit using Gradient Descent")
    plt.legend()
    plt.show()

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

theta = gradient_descent(X, y, learning_rate=0.1, n_iterations=1000)

print(f"Final weights (theta): \n{theta}")

plot_regression(X, y, theta)
```

```python
def compute_cost(X_b, y, theta):
   
    m = len(y)
    predictions = X_b.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

def gradient_descent(X, y, learning_rate=0.01, n_iterations=1000):
    m = len(X)  
    X_b = np.c_[np.ones((m, 1)), X]  
    
    theta = np.random.randn(X_b.shape[1], 1)
    
    cost_history = []
    
    for iteration in range(n_iterations):
        predictions = X_b.dot(theta)
        
        gradients = (2 / m) * X_b.T.dot(predictions - y)
        
        theta = theta - learning_rate * gradients
        
        cost = compute_cost(X_b, y, theta)
        cost_history.append(cost)
    
    return theta, cost_history

def plot_convergence(cost_history):

    plt.plot(cost_history)
    plt.xlabel("Iteration")
    plt.ylabel("Cost (Mean Squared Error)")
    plt.title("Cost Function Convergence")
    plt.show()

def plot_regression(X, y, theta):
    plt.scatter(X, y, color='blue', label='Data points')
    
    X_b = np.c_[np.ones((len(X), 1)), X]
    predicted_values = X_b.dot(theta)
    
    plt.plot(X, predicted_values, color='red', label='Predicted line')
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title("Linear Regression Fit using Gradient Descent")
    plt.legend()
    plt.show()

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

theta, cost_history = gradient_descent(X, y, learning_rate=0.1, n_iterations=1000)

print(f"Final weights (theta): \n{theta}")

plot_convergence(cost_history)

plot_regression(X, y, theta)
```
